# Total Available Market

For this project, I was asked to predict the total available market of electrical transformers in the Mexican private sector. To do this, I first had to merge seven tables to have all the required information in one dataframe. From there, I cleaned the data, homogenizing naming conventions and ensuring there were no outliers or nonsensical data points. I had to remove variables that didn't help predict the market size. The performance of the regression model was good, with the following values:

r2: 0.73

MAPE: 11.2%

For reference, the model was trying to predict the available market per month based on factors such as GDP, specific sector growth, etc. The estimate was made as units sold.
